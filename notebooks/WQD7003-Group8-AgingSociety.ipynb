## 1. Business Understanding

### 1.1 Background & Motivation

Malaysia is undergoing rapid population ageing, with the proportion of individuals aged 60 and above expected to increase significantly in the coming decades. This demographic shift places growing pressure on the healthcare system, as elderly individuals are major users of healthcare services. The high prevalence of non-communicable diseases such as diabetes, hypertension, and hyperlipidaemia further contributes to increased healthcare demand. Findings from the National Health and Morbidity Survey (NHMS) 2018 and 2019 highlight that chronic conditions and ageing are key drivers of healthcare utilisation in Malaysia.

### 1.2 Problem Statement

The increasing demand for healthcare services among the elderly highlights the need for a systematic and data-driven approach to identify individuals who are likely to require high levels of care. Without early identification of high-risk individuals, healthcare systems may face challenges in resource allocation, service planning, and timely intervention.

### 1.3 Project Objective

This project proposes a machine learning-based predictive model to identify elderly individuals with high healthcare demand. Using NHANES as a representative dataset and aligning selected variables with NHMS findings, the model aims to support the Ministry of Health Malaysia (MOH) in early risk detection, targeted intervention, and efficient healthcare resource planning.

### 1.4 Societal Contribution

This project contributes to society by supporting a shift from reactive to proactive healthcare. By identifying high-risk elderly individuals early, the model can assist healthcare providers in delivering preventive care, optimising resource allocation, and improving the overall quality of life for the ageing population.



## 2. Data Understanding

### 2.1 Data Source Overview

This study utilises data from the National Health and Nutrition Examination Survey (NHANES) 2021–2023 cycle, conducted by the Centers for Disease Control and Prevention (CDC), United States. NHANES is a nationally representative survey that collects comprehensive information on health status, healthcare utilisation, demographic characteristics, and lifestyle factors through interviews and clinical examinations.

The dataset is selected due to its rich individual-level health data, which supports the development of predictive models for healthcare demand. In addition, variables are aligned with key indicators identified in NHMS reports to ensure relevance to the Malaysian context.

### 2.2 Dataset Selection (to check and reconfirm)

To support the objective of predicting healthcare demand among the elderly, the following NHANES datasets are selected:

|**Dataset**|**Description**|
| :-: | :-- |
|DEMO|Demographic information (age, gender, income)|
|HUQ|Healthcare utilisation and access|
|BMX|Body measurements (BMI)|
|BPX|Blood pressure measurements|
|DIQ|Diabetes status|
|SMQ|Smoking behaviour|

These datasets provide a combination of demographic, clinical, and behavioural variables that are relevant for modelling healthcare demand.

### 2.3 Key Variables and Features (to check and reconfirm)

Based on the project objective and supported by NHMS findings, the following key features are identified:

**Demographic factors:** Age, gender, income
**Clinical factors:** Diabetes status, blood pressure, BMI
**Lifestyle factors:** Smoking behaviour
**Healthcare utilisation:** Frequency of healthcare visits (used to define target variable)

These variables represent important determinants of healthcare demand, particularly among elderly populations.

### 2.4 Data Characteristics

The NHANES dataset is cross-sectional and contains individual-level observations, where each respondent is uniquely identified by a respondent ID (SEQN). Data is collected through structured interviews and physical examinations, ensuring both subjective and objective health measurements.

The dataset may contain missing values and coded responses (e.g., “7” or “9” indicating unknown or refused responses), which will require appropriate data cleaning and preprocessing in the next phase.

### 2.5 Relevance to the Study

The selected datasets provide sufficient coverage of key factors influencing healthcare demand, including health conditions, lifestyle behaviours, and demographic characteristics. By aligning these variables with NHMS findings, the study ensures that the predictive model remains relevant to the Malaysian ageing population while leveraging the richness of NHANES data.

### 2.6 Initial Observations

Initial exploration of the NHANES dataset indicates that it contains a diverse range of demographic, clinical, and behavioural variables relevant to healthcare demand. Preliminary inspection shows variability in age distribution, health conditions, and lifestyle factors among respondents. Additionally, the dataset includes both self-reported and clinically measured variables, providing a comprehensive view of individual health status. Some variables contain missing or coded values, which will require data cleaning in the next phase. These observations suggest that the dataset is suitable for developing a predictive model to analyse healthcare demand among the elderly.
