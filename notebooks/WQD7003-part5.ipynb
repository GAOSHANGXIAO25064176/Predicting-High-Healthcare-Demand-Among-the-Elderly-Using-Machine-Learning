## Part 5: Model Evaluation and Result Interpretation

### 5.1 Definition of Health Risk Score

This study aims to construct a health risk scoring model for the elderly to identify individuals with high healthcare demand. The Health Risk Score is defined as the total number of the following nine health risk factors satisfied by an individual, each counted as 1 point, with a total score ranging from **0 to 9**:

- Diabetes
- Hypertension
- Hypercholesterolemia
- Arthritis
- Cardio_disease
- Chronic Obstructive Pulmonary Disease (COPD)
- Cancer
- Fair/poor health
- Functional problem

A higher risk score indicates a higher potential healthcare need.

According to the study design, elderly individuals who meet **2 or more risk factors (risk score ≥ 2)** are defined as the high health risk / high healthcare demand group.

This study uses three regression models to predict the continuous risk score, and then converts the score into a binary classification result based on a threshold (≥2) for high-risk identification.

### 5.2 Model Evaluation Metrics

To comprehensively evaluate model performance, both regression and classification metrics are adopted.

#### (1) Regression Metrics (predicting risk score 0–9)

- **MAE (Mean Absolute Error)**: Average absolute difference between predicted and true values; lower is better.
- **RMSE (Root Mean Squared Error)**: More sensitive to large errors; lower is better.
- **R² (Coefficient of Determination)**: Proportion of variance in the target variable explained by the model; closer to 1 indicates better fit.

#### (2) Classification Metrics (high-risk threshold: ≥2)

- Accuracy
- Precision
- Recall
- F1-score
- ROC-AUC

Since the proportion of the high-risk group may be relatively high, F1-score and AUC better reflect the model's discriminative ability than accuracy alone.

This study constructed and compared three machine learning models to predict health risk scores for the elderly and identify those with high healthcare demand.
