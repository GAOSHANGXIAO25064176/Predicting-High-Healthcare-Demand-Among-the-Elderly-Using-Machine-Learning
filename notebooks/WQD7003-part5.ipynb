### 5.1 Target Variable and Evaluation Metrics

The goal of this study is to predict the healthcare demand group for elderly respondents aged 60 and above:

- **target = 0**: Lower healthcare demand group (number of healthcare visits in the past year < 4)
- **target = 1**: Higher healthcare demand group (number of healthcare visits in the past year ≥ 4)

According to the project requirements, the final model should accurately identify elderly individuals with high healthcare demand to facilitate resource allocation and early intervention.

#### Evaluation Metrics (Binary Classification)

- **Accuracy**: Proportion of correctly predicted samples.
- **Precision**: Among samples predicted as positive (high demand), the proportion that are actually positive.
- **Recall**: Among actual positive samples, the proportion correctly predicted.
- **F1 Score**: Harmonic mean of precision and recall, providing a balanced measure.
- **AUC (Area Under the ROC Curve)**: Measures the model's ability to distinguish between positive and negative classes; closer to 1 is better.
- **Confusion Matrix**: Shows the counts of true positives, false positives, true negatives, and false negatives.

Because the proportions of high and low healthcare demand may be imbalanced, F1 score and AUC better reflect the model's true performance than accuracy alone.

### 5.7 Conclusion

This study, based on NHANES 2017–2020 data, constructed three machine learning models to predict healthcare demand groups (lower vs. higher) for elderly individuals aged 60 and above.

- **Logistic Regression** provides an interpretable baseline, confirming that cardiovascular disease, diabetes, and age are key factors positively associated with high healthcare demand.
- **Random Forest** improves prediction accuracy while outputting feature importance, clearly showing which health issues most contribute to high healthcare demand, thus providing data support for policy interventions.
- **XGBoost** achieves the best performance (AUC = 0.88) and can be used as the final model for screening high‑risk elderly individuals in real‑world scenarios.

Overall, ensemble learning methods demonstrate clear advantages in handling complex health data and identifying high healthcare demand groups.
