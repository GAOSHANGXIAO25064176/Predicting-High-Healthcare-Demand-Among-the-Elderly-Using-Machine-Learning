### 1.1 Business Understanding

#### 1.1.1 Background & Motivation
Malaysia is experiencing a rapidly aging population, with the proportion of individuals aged 65 and above expected to increase significantly in the coming years. This demographic shift places increasing pressure on the healthcare system, as older adults are more likely to suffer from chronic diseases such as hypertension, diabetes, and cardiovascular conditions.

In addition, aging is often associated with functional decline, including limitations in mobility and daily activities. These health and functional issues contribute to higher healthcare utilization, including more frequent doctor visits, hospitalizations, and long-term care needs.

Findings from the National Health and Morbidity Survey (NHMS) 2018 and 2019 highlight that chronic conditions and ageing are key drivers of healthcare utilisation in Malaysia. Given these challenges, there is a growing need to better understand and manage healthcare demand among the elderly population.

#### 1.1.2 Problem Statement
Despite the increasing healthcare needs of older adults, many healthcare systems still operate in a reactive manner, responding to illness after it occurs rather than preventing it in advance. A key challenge is the **lack of effective methods to identify elderly individuals who are at high health risk or likely to have high healthcare demand**. Without early identification, healthcare providers may face difficulties in resource allocation, timely intervention, and preventive care planning.The increasing demand for healthcare services among the elderly highlights the need for a systematic and data-driven approach to identify individuals who are likely to require high levels of care. 

#### 1.1.3 Project Objective

This project proposes a **machine learning-based predictive model to identify elderly individuals with high healthcare demand**. Using NHANES as a representative dataset and aligning selected variables with NHMS findings, the model aims to support the Ministry of Health Malaysia (MOH) in early risk detection, targeted intervention, and efficient healthcare resource planning.
Specifically, the project aims to:
- Identify elderly individuals who are likely to have high healthcare demand
- Analyze key contributing factors such as demographic characteristics, chronic diseases, and functional status
- Support data-driven decision-making in healthcare planning

#### 1.1.4 Societal Contribution
This study has practical implications for public health and healthcare management.By identifying high-risk elderly individuals early, healthcare providers can implement preventive measures and targeted interventions. This can help **reduce disease progression, lower healthcare costs, and improve patient outcomes**. Furthermore, the model can **assist healthcare providers in delivering preventive care, optimising resource allocation, and improving the overall quality of life for the ageing population**.



### 1.2. Data Understanding

#### 1.2.1 Data Source Overview

This study uses data from the National Health and Nutrition Examination Survey (NHANES) 2017-March 2020 Pre-pandemic cycle. NHANES is a large-scale health survey conducted by the Centers for Disease Control and Prevention (CDC) in the United States. It is designed to provide nationally representative information on the health and nutritional status of the population.

One of the main reasons for selecting NHANES is that it contains a wide range of individual-level information, including demographic background, diagnosed medical conditions, health behaviours, access to healthcare, and cardiovascular- and diabetes-related indicators. These dimensions are highly relevant to the objective of this project, which is to predict healthcare demand among older adults.

Another strength of NHANES is that the data is collected through multiple sources, including household interviews, questionnaires, and physical examinations. This makes the dataset more reliable than a purely self-reported survey. For this project, NHANES provides a suitable data foundation because healthcare demand among the elderly is usually influenced by multiple interacting factors rather than by a single condition alone.

Although NHANES is a U.S.-based dataset, it is still useful for this study because the relationships between ageing, chronic disease burden, lifestyle factors, and healthcare utilization are broadly relevant in many settings. In addition, the variables available in NHANES are consistent with commonly used indicators in ageing and public health research. In addition, variables are aligned with key indicators identified in NHMS reports to ensure relevance to the Malaysian context.

#### 1.2.2 Dataset Selection   

To support the objective of predicting high healthcare demand among elderly individuals, seven NHANES datasets were selected. Each file contributes a different aspect of information needed for the analysis:

|**Dataset**|**Description**|
| :-: | :-- |
|DEMO|Demographic information (age, gender, income, race/ethnicity)|
|HUQ|Healthcare utilisation and access|
|BPQ|Blood pressure and hypertension-related information|
|DIQ|Diabetes-related information|
|SMQ|Smoking behaviour|
|MCQ|Medical conditions and chronic disease history|
|CDQ|Cardiovascular health-related information|

These datasets were chosen because healthcare demand among older adults is influenced by a combination of demographic, clinical, behavioural, and healthcare access factors. Using only one dataset would not be sufficient to capture the full picture. Therefore, the study combines several files to form a more complete analytical dataset. The datasets were merged using **SEQN, the unique respondent identifier in NHANES**. This merging process allows records from different survey components to be linked to the same participant.


#### 1.2.3 Key Variables and Features 

Based on the project objective and supported by NHMS findings, the following key features are identified:

**Demographic factors:**   
Demographic factors include age, gender, race/ethnicity, and income level. These variables are important because healthcare demand may vary across different social and population groups. Based on the NHMS 2019 report, the elderly are identified as **individuals aged 60 years old and above**.

**Clinical factors:**   
Chronic disease factors include conditions such as hypertension, diabetes, cardiovascular symptoms or history, and other medical conditions. These variables reflect the health burden faced by elderly respondents.

Blood pressure and diabetes factors focus more specifically on diagnosis and treatment status, such as whether a respondent has been told they have high blood pressure or diabetes, and whether they are taking related medication.

**Healthcare utilisation:**   
Healthcare utilization factors include variables related to healthcare access, doctor visits, and contact with health services. These are especially relevant because they are closely connected to the idea of healthcare demand. This data is used to support target variable construction.

**Lifestyle factors:**    
Lifestyle factors such as smoking behaviour are also included because they may be associated with poorer health outcomes and higher healthcare needs among older adults.

These variables represent important determinants of healthcare demand, particularly among elderly populations. They were selected because healthcare demand among the elderly is usually influenced by both medical conditions and healthcare access, rather than by a single factor alone.

#### 1.2.4 Data Characteristics

The combined dataset has several important characteristics that affect later analysis.

**Data includes both categorical and numerical variables**   
Many NHANES questionnaire variables are categorical because responses are recorded using coded answer options such as “Yes”, “No”, “Refused”, or “Don’t know”. At the same time, some demographic measures such as age and income ratio are numerical.

**Dataset is cross-sectional**   
The dataset meaning that captures respondents’ information at one point in time rather than following them over multiple years. This means the project is suitable for identifying associations and building predictive models based on current status, but it cannot directly establish long-term causal effects.

**Data comes from multiple files**   
The data structure is multi-source and modular. This is useful because it allows the study to include different dimensions of elderly health, but it also creates the need for proper merging, variable selection, and consistency checking.

**Dataset includes both self-reported measures and health-related indicators**   
This characteristic of the dataset provides a broader view of the elderly population. However, self-reported data may still be subject to recall bias or reporting errors, especially for older respondents.

The dataset may contain missing values and coded responses (e.g., “7” or “9” indicating unknown or refused responses), which will require appropriate data cleaning and preprocessing in the next phase.

#### 1.2.5 Relevance to the Study

The selected datasets are relevant to this study because they cover the main dimensions related to healthcare demand among older adults. Elderly individuals with similar ages may still differ greatly in chronic disease burden, lifestyle risk, and healthcare access. By combining demographic, medical, behavioural, utilization, and cardiovascular-related information, the datasets provide a broader view of their healthcare needs. By aligning these variables with NHMS findings, the study ensures that the predictive model remains relevant to the Malaysian ageing population while leveraging the richness of NHANES data.

#### 1.2.6 Initial Observations

Initial exploration of the NHANES dataset indicates that it contains a diverse range of demographic, clinical, behavioural, and healthcare-related variables relevant to healthcare demand. Preliminary inspection suggests variation in age, health conditions, smoking behaviour, and healthcare access among respondents. Some variables are likely to contain missing values or coded nonresponse categories, which will require data cleaning in the next phase. In addition, the dataset combines both self-reported and structured health-related information, making it suitable for developing a predictive model to analyse healthcare demand among the elderly.



### 1.3 Data Preparation - In Progress

The data preparation phase is currently in progress, involving the construction of the dataset using selected NHANES 2017–2020 dataset files: `DEMO`, `MCQ`, `BPQ`, `DIQ`, `HUQ`, `CDQ`, and `SMQ`. These files provide information on demographic characteristics, chronic diseases, blood pressure and cholesterol conditions, healthcare utilization, chest pain and cardiovascular symptoms, and smoking-related behaviours. These datasets are being merged using the unique respondent identifier (SEQN) to form a unified dataset for analysis.

The dataset is being filtered to focus on elderly individuals aged 60 years and above, in line with the objective of the study. At the same time, relevant variables are being selected based on NHMS findings, ensuring that key factors such as demographic characteristics, chronic conditions, lifestyle behaviours, and functional status are appropriately represented.

Data cleaning is also being carried out to address missing, inconsistent, and coded values (such as “Refused” or “Don’t know”), in order to improve data quality and reliability for subsequent analysis.

In addition, a target variable representing healthcare demand is being defined using selected indicators from the dataset. This will support the classification of individuals into different levels of healthcare demand for the modelling stage.


### 1.4 Modelling - Planned

The modelling phase will focus on developing predictive models to classify elderly individuals based on their level of healthcare demand, **lower healthcare needs group (target=0)** or **higher healthcare needs group (target=1)**. Selected features from the data preparation stage, including demographic, clinical, lifestyle, and functional variables, will be used as inputs for the models.

Several machine learning algorithms are planned to be explored, including **Logistic Regression**, **Decision Tree**, and **Random Forest**. These models are chosen for their ability to handle classification problems and to provide interpretable insights into the factors influencing healthcare demand.

The modelling process will involve splitting the dataset into training and testing sets, followed by model training and tuning. This will allow for a fair assessment of model performance and generalisation.

Overall, this phase aims to identify the most suitable model for predicting healthcare demand, while also highlighting the key variables that contribute to higher healthcare utilisation among the elderly.



### 1.5 Evaluation - Planned

The evaluation phase will assess the performance of the developed models in predicting healthcare demand among the elderly. Appropriate evaluation metrics such as accuracy, precision, recall, and confusion matrix will be used to measure the effectiveness of each model.

The results from different models will be compared to identify the most suitable approach for this study. In addition to performance metrics, the interpretability of the models will also be considered, particularly in understanding the key factors that influence healthcare demand.

This phase will ensure that the selected model is both reliable and meaningful, providing insights that align with the objectives of the study. The findings will also be interpreted in relation to NHMS insights to ensure relevance to the Malaysian healthcare context.

### 1.6 Deployment - Planned

The deployment phase will focus on translating the developed predictive model into practical insights that can support healthcare decision-making. The model is intended to assist in identifying elderly individuals with higher healthcare demand, enabling more targeted and proactive interventions.

Although full system implementation is beyond the scope of this study, the proposed model can be integrated into healthcare planning systems to support the Ministry of Health Malaysia (MOH) in resource allocation and service planning.

Overall, this phase highlights the potential real-world application of the model, contributing towards a more proactive and data-driven approach in managing healthcare demand among the ageing population.
